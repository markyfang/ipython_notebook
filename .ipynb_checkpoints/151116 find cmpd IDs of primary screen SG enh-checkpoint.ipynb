```python
#cell type dictionary keys
HEK293XT = 1
CVB_SMNPC = 2

#screen dictionary keys
UCSF_GENEN = 1
FDA_IN = 2
SPECTRUM = 3
LOPAC_PRESTWICK = 4
TPIMS = 5
```

```python
#path of the folder containing the data
analyses_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Data - analyses/SG project/screening/analyses of large screens/find compound IDs/'

#the {} is replaced by the argument in the string function .format()
#header = None means do NOT take the first row as column names for the dataframe being created by pd.read_csv
potential_SG_enh = pd.read_csv('{}/150722 SG enh RVM coord.csv'.format(analyses_folder), header = None)

#print out the default column names in the pandas dataFrame potential_hits
potential_SG_enh.columns.values

```

```python
#rename certain column names in the dataFrame potential_SG_enh
potential_SG_enh = potential_SG_enh.rename(columns = {3:'plate'})
potential_SG_enh = potential_SG_enh.rename(columns = {4:'row'})
potential_SG_enh = potential_SG_enh.rename(columns = {5:'column'})
potential_SG_enh = potential_SG_enh.rename(columns = {6:'cell type'})
potential_SG_enh = potential_SG_enh.rename(columns = {7:'screen'})

potential_SG_enh
```

```python
not_SG_enh = pd.read_csv('{}/151116 not actually SG enhancers.csv'.format(analyses_folder), header = None)

not_SG_enh.columns.values
```

```python

not_SG_enh = not_SG_enh.rename(columns = {6:'plate'})
not_SG_enh = not_SG_enh.rename(columns = {7:'row'})
not_SG_enh = not_SG_enh.rename(columns = {8:'column'})
not_SG_enh = not_SG_enh.rename(columns = {9:'cell type'})
not_SG_enh = not_SG_enh.rename(columns = {10:'screen'})
not_SG_enh = not_SG_enh.rename(columns = {11:'true SG enh'})


not_SG_enh
```

```python
not_SG_enh_coord = not_SG_enh.loc[:, 'plate':'true SG enh']
not_SG_enh_coord
not_SG_enh_coord.columns.values
```

```python
SG_and_not_SG_enh = pd.merge(potential_SG_enh, not_SG_enh_coord, how = 'outer', on = ['plate', 'row', 'column', 'cell type', 'screen'])
SG_and_not_SG_enh
```

```python
list(SG_and_not_SG_enh.columns.values)
true_SG_enh = SG_and_not_SG_enh.loc[SG_and_not_SG_enh.loc[:,'true SG enh'] != 0, :]
true_SG_enh = true_SG_enh.loc[true_SG_enh.loc[:, 'screen'] != 1, :]
true_SG_enh = true_SG_enh.loc[true_SG_enh.loc[:, 'screen'] != 5, :]
true_SG_enh.shape
```

```python
metadata_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Metadata/SG project/SG screening metadata/screen platemaps with plate, row, column info'
```

```python
fda_in_metadata = pd.read_csv('{}/fda-in.csv'.format(metadata_folder))
fda_in_metadata.head()
```

```python
spectrum_metadata = pd.read_csv('{}/spectrum.csv'.format(metadata_folder))
spectrum_metadata.head()
```

```python
lopac_metadata = pd.read_csv('{}/lopac.csv'.format(metadata_folder))
lopac_metadata.head()
```

```python
prestwick_metadata = pd.read_csv('{}/prestwick.csv'.format(metadata_folder))
prestwick_metadata.head()
```

```python
fda_in_id = pd.merge(fda_in_metadata, true_SG_enh, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
fda_in_id
```

```python
spectrum_id = pd.merge(spectrum_metadata, true_SG_enh, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
spectrum_id
```

```python
lopac_id = pd.merge(lopac_metadata, true_SG_enh, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
lopac_id
```

```python
prestwick_id = pd.merge(prestwick_metadata, true_SG_enh, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
prestwick_id
```

```python
with open ('{}/fda in SG enh IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    fda_in_id.to_csv('{}/fda in SG enh IDs.csv'.format(analyses_folder))
    
with open ('{}/spectrum SG enh IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    spectrum_id.to_csv('{}/spectrum SG enh IDs.csv'.format(analyses_folder))
    
with open ('{}/lopac SG enh IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    lopac_id.to_csv('{}/lopac SG enh IDs.csv'.format(analyses_folder))
    
with open ('{}/prestwick SG enh IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    prestwick_id.to_csv('{}/prestwick SG enh IDs.csv'.format(analyses_folder))
```
