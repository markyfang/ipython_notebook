```python
import numpy as np
import csv
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pylab
import scipy
import scipy.stats
import math
import pandas as pd
```

```python
#cell type dictionary keys
HEK293XT = 1
CVB_SMNPC = 2

#screen dictionary keys
UCSF_GENEN = 1
FDA_IN = 2
SPECTRUM = 3
LOPAC_PRESTWICK = 4
TPIMS = 5
```

```python
analyses_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Data - analyses/SG project/screening/analyses of large screens/150724 find hit compound IDs/'

#the {} is replaced by the argument in the string function .format()
#header = None means do NOT take the first row as column names for the dataframe being created by pd.read_csv
potential_hits = pd.read_csv('{}/150722 hits RVM coord.csv'.format(analyses_folder), header = None)

potential_hits.columns.values

```

```python

potential_hits = potential_hits.rename(columns = {3:'plate'})
potential_hits = potential_hits.rename(columns = {4:'row'})
potential_hits = potential_hits.rename(columns = {5:'column'})
potential_hits = potential_hits.rename(columns = {6:'cell type'})
potential_hits = potential_hits.rename(columns = {7:'screen'})

potential_hits
```

```python
false_pos = pd.read_csv('{}/150724 false positives.csv'.format(analyses_folder), header = None)

false_pos.columns.values
```

```python

false_pos = false_pos.rename(columns = {6:'plate'})
false_pos = false_pos.rename(columns = {7:'row'})
false_pos = false_pos.rename(columns = {8:'column'})
false_pos = false_pos.rename(columns = {9:'cell type'})
false_pos = false_pos.rename(columns = {10:'screen'})
false_pos = false_pos.rename(columns = {11:'true pos'})


false_pos
```

```python
false_pos_coord = false_pos.loc[:, 'plate':'true pos']
false_pos_coord
false_pos_coord.columns.values
```

```python
true_and_false_pos = pd.merge(potential_hits, false_pos_coord, how = 'outer', on = ['plate', 'row', 'column', 'cell type', 'screen'])
true_and_false_pos
```

```python
list(true_and_false_pos.columns.values)
true_pos = true_and_false_pos.loc[true_and_false_pos.loc[:,'true pos'] != 0, :]
#true_and_false_pos.loc[true_and_false_pos.loc["true_pos"] != 0, :]
true_pos = true_pos.loc[true_pos.loc[:, 'screen'] != 1, :]
true_pos = true_pos.loc[true_pos.loc[:, 'screen'] != 5, :]
true_pos.shape

#true_pos
```

```python
metadata_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Metadata/SG project/SG screening metadata/screen platemaps with plate, row, column info'
```

```python
fda_in_metadata = pd.read_csv('{}/fda-in.csv'.format(metadata_folder))
fda_in_metadata.head()
```

```python
spectrum_metadata = pd.read_csv('{}/spectrum.csv'.format(metadata_folder))
spectrum_metadata.head()
```

```python
lopac_metadata = pd.read_csv('{}/lopac.csv'.format(metadata_folder))
lopac_metadata.head()
```

```python
prestwick_metadata = pd.read_csv('{}/prestwick.csv'.format(metadata_folder))
prestwick_metadata.head()
```

```python
fda_in_id = pd.merge(fda_in_metadata, true_pos, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
fda_in_id
```

```python
spectrum_id = pd.merge(spectrum_metadata, true_pos, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
spectrum_id
```

```python
lopac_id = pd.merge(lopac_metadata, true_pos, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
lopac_id
```

```python
prestwick_id = pd.merge(prestwick_metadata, true_pos, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
prestwick_id
```

```python
with open ('{}/fda in IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    fda_in_id.to_csv('{}/fda in IDs.csv'.format(analyses_folder))
    
with open ('{}/spectrum IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    spectrum_id.to_csv('{}/spectrum IDs.csv'.format(analyses_folder))
    
with open ('{}/lopac IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    lopac_id.to_csv('{}/lopac IDs.csv'.format(analyses_folder))
    
with open ('{}/prestwick IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    prestwick_id.to_csv('{}/prestwick IDs.csv'.format(analyses_folder))
```
