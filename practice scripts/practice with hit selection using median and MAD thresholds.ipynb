```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import scipy
import scipy.stats
```

```python
np.set_printoptions(precision = 4, suppress = True)

size = 384
arr = scipy.stats.norm.rvs(size = size)
arr = np.reshape(arr, (16, 24))
print(arr)
avg = np.average(arr)
sd = np.std(arr)
print(avg)
print(sd)

plate_median = np.median(arr)
print(plate_median)

mad = 1.4826 * np.median(np.absolute(arr - plate_median))
print(mad)

upper_threshold = plate_median + 2 * mad
#2 MAD corresponds to false negative rate of 0.023
lower_threshold = plate_median - 2 * mad

#rows = arr.shape[0]
#cols = arr.shape[1]
#print(num_row)
#print(num_col)

```

```python
hits = []
sg_enhancers = []
#print(hits)

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] < lower_threshold:
            #plate well coordinates of hits, indexed from 1 rather than 0
            hits.append(i + 1)
            hits.append(j + 1)
            #hits.append((i + 1, j + 1))
        if arr[i][j] > upper_threshold:
            sg_enhancers.append(i + 1)
            sg_enhancers.append(j + 1)


len_hits = len(hits)
#reshape into ordered pairs giving the plate well coordinates for hits
hits = np.reshape(hits, (len_hits / 2, 2))
print(hits)
len_sg_enhancers = len(sg_enhancers)
sg_enhancers = np.reshape(sg_enhancers, (len_sg_enhancers / 2, 2))
print(sg_enhancers)
            
```
