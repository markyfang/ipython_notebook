```python
#this ensures that plots produced by matplotlib (and seaborn, which is built off of matplotlib) are displayed inline in the ipython interface, rather than opening a new window
%matplotlib inline

import numpy as np
import csv
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pylab
import scipy
import scipy.stats
import math
import pandas as pd
import re
import seaborn as sns
```

```python
#path of the folder containing the tpims data
tpims_data_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Data - analyses/SG project/screening/analyses of large screens/150701 sg results'

#the {} is replaced by the argument in the string function .format()
tpims_hits = pd.read_csv('{}/151205 tpims hits.csv'.format(tpims_data_folder), header = None)

tpims_sg_enh = pd.read_csv('{}/151205 tpims SG enh.csv'.format(tpims_data_folder), header = None)
```

```python
tpims_hits = tpims_hits.rename(columns = {0:'plate'})
tpims_hits = tpims_hits.rename(columns = {1:'row'})
tpims_hits = tpims_hits.rename(columns = {2:'column'})
tpims_hits = tpims_hits.rename(columns = {3:'cell type'})
tpims_hits = tpims_hits.rename(columns = {4:'count'})

tpims_hits
```

```python
#find compounds which came up as hits across multiple concentrations/dilutions
tpims_hits_mult_conc = tpims_hits.groupby(['row', 'column', 'cell type'])['count'].sum()

tpims_hits_mult_conc = tpims_hits_mult_conc.to_frame()

tpims_hits_mult_conc = tpims_hits_mult_conc.reset_index(drop = False)

tpims_hits_mult_conc
```

```python
#find compounds which came up as hits across multiple cell types
tpims_hits_mult_celltype = tpims_hits.groupby(['plate', 'row', 'column'])['count'].sum()

tpims_hits_mult_celltype = tpims_hits_mult_celltype.to_frame()

tpims_hits_mult_celltype = tpims_hits_mult_celltype.reset_index(drop = False)

tpims_hits_mult_celltype
```

```python
#find compounds which came up as hits across multiple conc and cell types
tpims_hits_mult_conc_and_celltype = tpims_hits.groupby(['row', 'column'])['count'].sum()

tpims_hits_mult_conc_and_celltype = tpims_hits_mult_conc_and_celltype.to_frame()

tpims_hits_mult_conc_and_celltype = tpims_hits_mult_conc_and_celltype.reset_index(drop = False)

tpims_hits_mult_conc_and_celltype
```

```python

tpims_sg_enh = tpims_sg_enh.rename(columns = {0:'plate'})
tpims_sg_enh = tpims_sg_enh.rename(columns = {1:'row'})
tpims_sg_enh = tpims_sg_enh.rename(columns = {2:'column'})
tpims_sg_enh = tpims_sg_enh.rename(columns = {3:'cell type'})
tpims_sg_enh = tpims_sg_enh.rename(columns = {4:'count'})

tpims_sg_enh
```

```python
#find compounds which came up as hits across multiple concentrations/dilutions
tpims_sg_enh_mult_conc = tpims_sg_enh.groupby(['row', 'column', 'cell type'])['count'].sum()

tpims_sg_enh_mult_conc = tpims_sg_enh_mult_conc.to_frame()

tpims_sg_enh_mult_conc = tpims_sg_enh_mult_conc.reset_index(drop = False)

tpims_sg_enh_mult_conc
```

```python
#find compounds which came up as hits across multiple cell types
tpims_sg_enh_mult_celltype = tpims_sg_enh.groupby(['plate', 'row', 'column'])['count'].sum()

tpims_sg_enh_mult_celltype = tpims_sg_enh_mult_celltype.to_frame()

tpims_sg_enh_mult_celltype = tpims_sg_enh_mult_celltype.reset_index(drop = False)

tpims_sg_enh_mult_celltype
```

```python
#find compounds which came up as hits across multiple conc and cell types
tpims_sg_enh_mult_conc_and_celltype = tpims_sg_enh.groupby(['row', 'column'])['count'].sum()

tpims_sg_enh_mult_conc_and_celltype = tpims_sg_enh_mult_conc_and_celltype.to_frame()

tpims_sg_enh_mult_conc_and_celltype = tpims_sg_enh_mult_conc_and_celltype.reset_index(drop = False)

tpims_sg_enh_mult_conc_and_celltype
```
