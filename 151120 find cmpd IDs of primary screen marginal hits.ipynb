```python
#path of the folder containing the data
analyses_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Data - analyses/SG project/screening/analyses of large screens/find compound IDs'

#the {} is replaced by the argument in the string function .format()
#header = None means do NOT take the first row as column names for the dataframe being created by pd.read_csv
marginal_hits = pd.read_csv('{}/151116 marginal hits.csv'.format(analyses_folder), header = None)

#print out the default column names in the pandas dataFrame potential_hits
marginal_hits.columns.values
marginal_hits

```

```python
#rename certain column names in the dataFrame marginal_hits
marginal_hits = marginal_hits.rename(columns = {6:'plate'})
marginal_hits = marginal_hits.rename(columns = {7:'row'})
marginal_hits = marginal_hits.rename(columns = {8:'column'})
marginal_hits = marginal_hits.rename(columns = {9:'cell type'})
marginal_hits = marginal_hits.rename(columns = {10:'screen'})

marginal_hits
```

```python
metadata_folder = '/Users/markfang/Dropbox/UCSD Grad work/RNA-Yeo Lab/Metadata/SG project/SG screening metadata/screen platemaps with plate, row, column info'
```

```python
fda_in_metadata = pd.read_csv('{}/fda-in.csv'.format(metadata_folder))
fda_in_metadata.head()
```

```python
spectrum_metadata = pd.read_csv('{}/spectrum.csv'.format(metadata_folder))
spectrum_metadata.head()
```

```python
lopac_metadata = pd.read_csv('{}/lopac.csv'.format(metadata_folder))
lopac_metadata.head()
```

```python
prestwick_metadata = pd.read_csv('{}/prestwick.csv'.format(metadata_folder))
prestwick_metadata.head()
```

```python
fda_in_id = pd.merge(fda_in_metadata, marginal_hits, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
fda_in_id
```

```python
spectrum_id = pd.merge(spectrum_metadata, marginal_hits, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
spectrum_id
```

```python
lopac_id = pd.merge(lopac_metadata, marginal_hits, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
lopac_id
```

```python
prestwick_id = pd.merge(prestwick_metadata, marginal_hits, how = 'inner', on = ['plate', 'row', 'column', 'screen'])
prestwick_id
```

```python
with open ('{}/fda in marginal hit IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    fda_in_id.to_csv('{}/fda in marginal hit IDs.csv'.format(analyses_folder))
    
with open ('{}/spectrum marginal hit IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    spectrum_id.to_csv('{}/spectrum marginal hit IDs.csv'.format(analyses_folder))
    
with open ('{}/lopac marginal hit IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    lopac_id.to_csv('{}/lopac marginal hit IDs.csv'.format(analyses_folder))
    
with open ('{}/prestwick marginal hit IDs.csv'.format(analyses_folder),'w') as id_outputfile:
    prestwick_id.to_csv('{}/prestwick marginal hit IDs.csv'.format(analyses_folder))
```
